```python
documents = [
    "Machine learning is a field of artificial intelligence that focuses on training algorithms using data.",
    "Deep learning is a subset of machine learning that uses neural networks with many layers.",
    "Natural language processing enables computers to understand human language.",
    "Computer vision allows machines to interpret and analyze visual information from the world.",
    "Reinforcement learning is a type of machine learning where agents learn by interacting with environments."
]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

print("Model loaded successfully!")
```

```python
import numpy as np

# Generate embeddings for documents
embeddings = model.encode(documents)

print("Embedding shape:", embeddings.shape)
```

```python
import requests
import json

BASE_URL = "http://localhost:8080/api/v1"

for i, vector in enumerate(embeddings):
    payload = {
        "id": f"doc_{i}",
        "values": vector.tolist(),
        "metadata": {
            "text": documents[i]
        }
    }

    response = requests.post(
        f"{BASE_URL}/indexes/documents/documents",
        json=payload
    )

    print(f"Inserted doc_{i}:", response.status_code)
```
